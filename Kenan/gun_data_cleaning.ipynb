```python
# Dependencies and Setup
import pandas as pd

# Study data files
guns_data_path = "./guns.csv"

# Read the data 
guns_data = pd.read_csv(guns_data_path)
```

```python
guns_data.columns
```

```python
guns_clean_df = guns_data.drop(['address', 'n_killed', 'n_injured', 'incident_url', 'source_url', 'incident_url_fields_missing', 'congressional_district', 'gun_stolen',
       'gun_type','location_description','n_guns_involved', 'notes', 'participant_age', 'participant_age_group', 'participant_gender',
       'participant_name', 'participant_relationship', 'participant_status',
       'participant_type', 'sources', 'state_house_district',
       'state_senate_district'  ], axis=1)
guns_clean_df = guns_clean_df.set_index('incident_id')
```

```python
guns_clean_df.head()
```

```python
guns_clean_no_incident_df = guns_clean_df.drop('incident_characteristics', axis=1)
guns_clean_no_incident_df.to_csv('guns_clean.csv')
```

```python
# Split the column into multiple columns
split_columns = guns_clean_df['incident_characteristics'].str.split('\|\|', expand=True)
num_columns = split_columns.shape[1]

# Assign split columns to the original DataFrame
guns_clean_df[list(range(num_columns))] = split_columns

# Display the updated DataFrame
guns_clean_df.head()
```

```python
guns_clean_df['date'] = pd.to_datetime(guns_clean_df['date'])
```

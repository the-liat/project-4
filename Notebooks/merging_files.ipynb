```python
import pandas as pd
```

```python
# Read the main file
mass_shooting = pd.read_csv('../Resources/mass_shooting_1982to2023_cleaned.csv')
set(mass_shooting['state'].to_list())
```

```python
# need to clean trailing spaces from state
mass_shooting['state'] = mass_shooting['state'].str.strip() 
mass_shooting
```

```python
# Read the gun laws file
gun_laws = pd.read_csv('../Resources/gun_laws_1991to2019_cleaned.csv')
gun_laws
```

```python
# Merge the mass shooting file with the gun laws file based on 'year' and 'state'
merged_file = pd.merge(mass_shooting, gun_laws, left_on=['year', 'state'], right_on=['YEAR', 'STATE'], how='outer')
merged_file
```

```python
# Read the demographics file
demographics = pd.read_csv('../Resources/us_demogrpahics_2011to2019_cleaned.csv', encoding="utf-8")
demographics
```

```python
# Merge the merged file with the demographics file based on 'year' and 'state'
merged_file = pd.merge(merged_file, demographics, left_on=['year', 'state'], right_on=['Year', 'State'], how='outer')
merged_file
```

```python
# Save the merged file to a new CSV file
merged_file.to_csv('..\Resources\mass_shootings_merged_data.csv', index=False)
```
